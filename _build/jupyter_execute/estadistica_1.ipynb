### 📌 **Datos Cualitativos (Categóricos)**
Los datos cualitativos son aquellos que **describen atributos o características** de un conjunto de elementos, sin involucrar valores numéricos que puedan ser operados matemáticamente. Se utilizan para **clasificar** y **categorizar** la información. Dentro de los datos cualitativos, encontramos dos subcategorías principales:

#### 🔹 **Datos Nominales**  
Son aquellos que simplemente **nombran o etiquetan** características sin que exista un orden jerárquico entre ellas. En estos datos, no se puede decir que una categoría es mayor o menor que otra.

✅ **Ejemplos**:  
- Género (Masculino, Femenino, No binario).  
- Tipo de automóvil (Sedán, SUV, Hatchback).  
- País de origen (Colombia, Argentina, México).  
- Color de un producto (Rojo, Azul, Verde).  

> **📌 Nota**: Aunque algunas veces los datos nominales pueden representarse con números (Ejemplo: 1 para Rojo, 2 para Azul), esto no significa que puedan ser sometidos a operaciones matemáticas.

#### 🔹 **Datos Ordinales**  
En este caso, las categorías **tienen un orden lógico o jerárquico**, pero las diferencias entre los valores no son necesariamente uniformes o medibles.

✅ **Ejemplos**:  
- Nivel de satisfacción en una encuesta (Excelente, Bueno, Regular, Malo).  
- Clasificación de hoteles (1 estrella, 2 estrellas, 3 estrellas…).  
- Tamaño de ropa (S, M, L, XL).  
- Nivel educativo alcanzado (Primaria, Secundaria, Universitario, Posgrado).  

> **📌 Importante**: Aunque los datos ordinales pueden ordenarse, **no se pueden realizar operaciones matemáticas** directamente con ellos. Por ejemplo, la diferencia entre "Regular" y "Bueno" no es numéricamente equivalente a la diferencia entre "Bueno" y "Excelente".

---

### 🔢 **Datos Cuantitativos (Numéricos)**
Estos datos **representan cantidades o medidas** y permiten aplicar operaciones matemáticas como sumas, restas, promedios, etc. Se dividen en:

#### 🔹 **Datos Cuantitativos Discretos**  
Son aquellos que **se obtienen mediante conteo** y solo pueden tomar **valores enteros**. No pueden fraccionarse en decimales o valores intermedios.

✅ **Ejemplos**:  
- Número de empleados en una empresa.  
- Cantidad de productos vendidos en un día.  
- Número de hijos en una familia.  
- Veces que un cliente compra en una tienda al mes.  

> **📌 Característica clave**: Se obtienen por conteo y no admiten valores decimales. Por ejemplo, no podemos tener **2.5 empleados** en una empresa.

#### 🔹 **Datos Cuantitativos Continuos**  
Representan mediciones que pueden **tomar cualquier valor dentro de un rango determinado**, incluyendo valores fraccionarios o decimales.

✅ **Ejemplos**:  
- Peso de un producto (500.3 gramos).  
- Altura de una persona (1.75 metros).  
- Precio de una acción en la bolsa (34.78 USD).  
- Tiempo de espera en un servicio (2.35 minutos).  

> **📌 Característica clave**: Se obtienen a través de **mediciones** y pueden tomar valores decimales o fraccionarios.

---

### 📆 **¿Dónde ubicamos las fechas?**  
Las **fechas** son un tipo de dato particular que **puede clasificarse de diferentes maneras** dependiendo del contexto y el tipo de análisis que se realice.  

#### 🔹 **Fechas como datos cualitativos nominales**  
Cuando las fechas **se usan como etiquetas** sin un significado numérico, se consideran **cualitativas nominales**. 

✅ **Ejemplos**:  
- Años de producción de una empresa (2010, 2015, 2020, 2025).  
- Mes de nacimiento de los empleados (Enero, Febrero, Marzo).  
- Día de la semana con más ventas (Lunes, Martes, Miércoles).  

> **📌 Nota**: En este caso, no se pueden realizar cálculos como restas o promedios, ya que las fechas solo sirven como identificadores.

#### 🔹 **Fechas como datos cuantitativos continuos**  
Si se usan para **medir intervalos de tiempo**, como diferencias entre fechas o duración de eventos, se consideran **cuantitativas continuas**.

✅ **Ejemplos**:  
- Tiempo en días entre dos eventos importantes.  
- Edad de una persona (calculada a partir de la fecha de nacimiento).  
- Tiempo transcurrido desde el último mantenimiento de una máquina.  
- Duración de un contrato en años.  

> **📌 Importante**: Si las fechas se restan para encontrar intervalos, **se comportan como datos cuantitativos continuos**.

#### 🔹 **Fechas como datos ordinales**  
En algunos casos, si las fechas representan un orden, pero no se usan para cálculos de intervalos, pueden tratarse como **datos ordinales**.

✅ **Ejemplo**:  
- Clasificación de eventos por fecha (Primer, Segundo, Tercer lugar en una competencia).  
- Orden cronológico de lanzamientos de productos.  

---

### 🔄 **¿Cómo elegir la mejor clasificación?**  
Dependerá de **cómo se usen los datos** en el análisis estadístico:

✅ **Si las fechas sirven como etiquetas, son cualitativas nominales.**  
✅ **Si se usan para medir diferencias de tiempo, son cuantitativas continuas.**  
✅ **Si solo se ordenan cronológicamente sin medir diferencias, pueden ser ordinales.**


## 📌 **Ejemplo 1: Datos Cualitativos**
Vamos a trabajar con **datos nominales y ordinales**, creando un `DataFrame` y visualizándolos con gráficos.

````python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creamos un DataFrame con datos cualitativos nominales y ordinales
df_cualitativo = pd.DataFrame({
    "Cliente": ["Carlos", "Ana", "Pedro", "Sofía", "Luis"],
    "Color_Favorito": ["Rojo", "Azul", "Verde", "Azul", "Rojo"],  # Nominal
    "Satisfacción": ["Bueno", "Excelente", "Regular", "Malo", "Bueno"]  # Ordinal
})

# Mapeo para convertir los datos ordinales a valores numéricos
orden_satisfaccion = {"Malo": 1, "Regular": 2, "Bueno": 3, "Excelente": 4}
df_cualitativo["Satisfacción_Num"] = df_cualitativo["Satisfacción"].map(orden_satisfaccion)

# Mostrar datos
display(df_cualitativo)

# Visualización de datos cualitativos nominales con un conteo de colores favoritos
plt.figure(figsize=(6, 4))
sns.countplot(data=df_cualitativo, x="Color_Favorito", palette="Set2")
plt.title("Distribución de Colores Favoritos")
plt.xlabel("Color Favorito")
plt.ylabel("Frecuencia")
plt.show()

# Visualización de datos ordinales con una barra de satisfacción
plt.figure(figsize=(6, 4))
sns.barplot(data=df_cualitativo, x="Cliente", y="Satisfacción_Num", palette="coolwarm")
plt.title("Nivel de Satisfacción de Clientes")
plt.ylabel("Nivel de Satisfacción (Ordinal)")
plt.xlabel("Cliente")
plt.show()
````

## 🔢 **Ejemplo 2: Datos Cuantitativos Discretos**
En este caso, trabajaremos con **número de compras de clientes** en una tienda (un dato discreto).

````python
# Creación de un DataFrame con datos discretos
df_discreto = pd.DataFrame({
    "Cliente": ["Carlos", "Ana", "Pedro", "Sofía", "Luis"],
    "Compras_Mensuales": [5, 12, 8, 15, 9]  # Discreto
})

# Mostrar datos
display(df_discreto)

# Visualización de las compras mensuales por cliente
plt.figure(figsize=(6, 4))
sns.barplot(data=df_discreto, x="Cliente", y="Compras_Mensuales", palette="viridis")
plt.title("Número de Compras Mensuales por Cliente")
plt.ylabel("Compras")
plt.xlabel("Cliente")
plt.show()
````

## 📉 **Ejemplo 3: Datos Cuantitativos Continuos**
Aquí analizaremos **precios de productos** en una tienda.

````python
import numpy as np

# Generamos datos continuos simulando precios de productos
df_continuo = pd.DataFrame({
    "Producto": ["Laptop", "Celular", "Tablet", "Monitor", "Teclado"],
    "Precio": [1200.50, 800.99, 450.75, 300.49, 75.89]  # Continuo
})

# Mostrar datos
display(df_continuo)

# Visualización de precios con un diagrama de dispersión
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df_continuo, x="Producto", y="Precio", color="blue", s=100)
plt.title("Precios de Productos")
plt.ylabel("Precio en USD")
plt.xlabel("Producto")
plt.grid(True)
plt.show()
````

## 📆 **Ejemplo 4: Fechas como Datos Cuantitativos y Cualitativos**
Las fechas pueden tratarse de **diferentes maneras** según el análisis.

````python

# Creamos datos de fechas
df_fechas = pd.DataFrame({
    "Evento": ["Lanzamiento Producto A", "Lanzamiento Producto B", "Campaña Publicitaria"],
    "Fecha": pd.to_datetime(["2023-01-15", "2023-06-20", "2023-09-05"])  # Fechas
})

# Extraemos datos temporales
df_fechas["Año"] = df_fechas["Fecha"].dt.year  # Como Cualitativo
df_fechas["Días desde Inicio Año"] = (df_fechas["Fecha"] - pd.Timestamp("2023-01-01")).dt.days  # Como Cuantitativo

# Mostrar datos
display(df_fechas)

# Visualización de diferencias de fechas
plt.figure(figsize=(6, 4))
sns.lineplot(data=df_fechas, x="Días desde Inicio Año", y="Año", marker="o", linestyle="--", color="red")
plt.title("Diferencias en Fechas de Eventos")
plt.ylabel("Año")
plt.xlabel("Días desde Inicio del Año")
plt.grid(True)
plt.show()
````

## 🔍 **Inferencia Estadística**
Hasta ahora hemos hablado de la **estadística descriptiva**, que nos permite organizar y resumir datos. Sin embargo, muchas veces queremos **hacer conclusiones sobre una población sin analizar todos sus elementos**. Aquí es donde entra la **inferencia estadística**.

**📌 Definición**  
La inferencia estadística es el proceso de **extraer conclusiones sobre una población a partir de una muestra**. Se basa en la teoría de probabilidades y permite:

- **Estimaciones** de parámetros desconocidos (media, proporción, varianza).  
- **Pruebas de hipótesis** para evaluar afirmaciones sobre una población.  
- **Modelos predictivos** para tomar decisiones con datos limitados.  

```{admonition} 🎯 Ejemplo real
:class: tip
Supongamos que queremos saber la **edad promedio de los clientes de una tienda**. No podemos preguntar a todos, así que tomamos una **muestra** y usamos inferencia estadística para estimar la edad promedio de toda la población.
```

### 📊 **Ejemplo Práctico en Python: Inferencia Estadística**
En este ejemplo, usaremos una **muestra aleatoria** para estimar la media poblacional.

````python
import numpy as np
import scipy.stats as stats

# Simulamos una población de edades de clientes (en años)
np.random.seed(42)
poblacion_edades = np.random.normal(loc=35, scale=10, size=10000)  # Media 35, Desviación 10, 10,000 clientes

# Tomamos una muestra aleatoria de 100 clientes
muestra = np.random.choice(poblacion_edades, size=100, replace=False)

# Calculamos la media muestral y el intervalo de confianza del 95%
media_muestral = np.mean(muestra)
intervalo_confianza = stats.t.interval(0.95, len(muestra)-1, loc=media_muestral, scale=stats.sem(muestra))

# Resultados
print(f"📌 Media muestral estimada: {media_muestral:.2f} años")
print(f"✅ Intervalo de confianza del 95%: {intervalo_confianza}")
````

📌 **Interpretación**:  
El intervalo de confianza nos dice que, con un **95% de certeza**, la **media de la población** está dentro de ese rango. Si repitiéramos el experimento muchas veces, **el 95% de los intervalos calculados contendrían la media real**.

### 📊 **1. Tipos de Datos**
Los datos pueden clasificarse en **cualitativos** o **cuantitativos**, y dentro de estas categorías, en diferentes subtipos según su naturaleza y aplicación.

✅ **Datos Cualitativos (Categóricos)**: Son atributos o características que no pueden ser medidas numéricamente, pero sí pueden ser clasificadas.  
   - **Nominales**: No tienen orden jerárquico. Ejemplo: color de autos, género.  
   - **Ordinales**: Tienen un orden lógico, pero las diferencias entre categorías no son uniformes. Ejemplo: nivel de satisfacción, clasificación de hoteles.

✅ **Datos Cuantitativos (Numéricos)**: Representan medidas o conteos, permitiendo operaciones matemáticas.  
   - **Discretos**: Se obtienen por conteo y solo pueden tomar valores enteros. Ejemplo: número de empleados, cantidad de productos vendidos.  
   - **Continuos**: Se obtienen por medición y pueden tomar valores decimales. Ejemplo: peso de un producto, temperatura.

✅ **Fechas y su Clasificación**:  
   - **Nominales** cuando se agrupan en categorías (Ejemplo: Mes de nacimiento).  
   - **Ordinales** cuando establecen un orden cronológico.  
   - **Continuas** cuando se usan para calcular intervalos de tiempo (Ejemplo: edad de una persona).

```{admonition} 📌 Importante
:class: note
La correcta clasificación de los datos es esencial para elegir las herramientas estadísticas adecuadas.
```

---

### 📏 **2. Escalas de Medición**
Determinan cómo los datos pueden ser analizados e interpretados:

✅ **Nominal**: Solo categorizan, sin un orden específico. Ejemplo: marcas de autos.  
✅ **Ordinal**: Ordenan categorías sin cuantificar diferencias. Ejemplo: nivel de educación.  
✅ **Intervalo**: Permiten medir diferencias, pero sin un cero absoluto. Ejemplo: temperatura en °C.  
✅ **Razón**: Poseen un cero absoluto, permitiendo comparaciones proporcionales. Ejemplo: ingresos de una empresa.

```{admonition} 🎯 Clave para el Análisis
:class: tip
La escala de medición define qué operaciones matemáticas y estadísticas son apropiadas.
```

---

### 🔍 **3. Inferencia Estadística**
Nos permite extraer conclusiones sobre una **población completa** a partir de una **muestra**.

✅ **Estimación de Parámetros**:  
   - La **media muestral** se usa para estimar la media poblacional.  
   - Los **intervalos de confianza** indican un rango probable para la verdadera media.

✅ **Pruebas de Hipótesis**:  
   - Se formulan afirmaciones sobre una población y se evalúan estadísticamente.  
   - Ejemplo: ¿Un nuevo medicamento reduce la presión arterial?
