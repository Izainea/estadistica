## üìä Definici√≥n Matem√°tica del Modelo Lineal

Un **modelo de regresi√≥n lineal m√∫ltiple** con $p$ variables predictoras se define como:

$$
Y_i = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \dots + \beta_p X_{ip} + \varepsilon_i
$$

En notaci√≥n matricial:

$$
\mathbf{Y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\varepsilon}
$$

donde:

- $\mathbf{Y} \in \mathbb{R}^{n \times 1}$ es el vector de respuestas.
- $\mathbf{X} \in \mathbb{R}^{n \times (p+1)}$ es la **matriz de dise√±o** con una columna de unos para el intercepto.
- $\boldsymbol{\beta} \in \mathbb{R}^{(p+1) \times 1}$ es el vector de coeficientes.
- $\boldsymbol{\varepsilon} \in \mathbb{R}^{n \times 1}$ representa los errores aleatorios.

### üîé ¬øQu√© buscamos en el modelo?
- **Estimaciones de los coeficientes** $\boldsymbol{\beta}$ que minimicen los errores.
- **Validaci√≥n de los supuestos** que garanticen la validez del modelo.

Pasemos a los **supuestos fundamentales** del modelo lineal.

# üìå Estimaci√≥n de Par√°metros en Modelos Lineales

La estimaci√≥n de los par√°metros de un **modelo de regresi√≥n lineal m√∫ltiple** se realiza mediante el **m√©todo de m√≠nimos cuadrados ordinarios (OLS, por sus siglas en ingl√©s)**. 

## ‚úÖ Planteamiento del Problema

Dado un conjunto de datos con $n$ observaciones y $p$ variables predictoras, podemos expresar el modelo en forma matricial como:

$$
\mathbf{Y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\varepsilon}
$$

donde:

- $\mathbf{Y} \in \mathbb{R}^{n \times 1}$ es el vector de respuestas (variable dependiente).
- $\mathbf{X} \in \mathbb{R}^{n \times (p+1)}$ es la **matriz de dise√±o**, donde cada fila representa una observaci√≥n y cada columna una variable predictora. Se incluye una columna de unos para representar el t√©rmino intercepto.
- $\boldsymbol{\beta} \in \mathbb{R}^{(p+1) \times 1}$ es el vector de coeficientes a estimar.
- $\boldsymbol{\varepsilon} \in \mathbb{R}^{n \times 1}$ es el vector de errores aleatorios.

La matriz de dise√±o $\mathbf{X}$ tiene la siguiente estructura:

$$
\mathbf{X} =
\begin{bmatrix}
1 & X_{11} & X_{12} & \dots & X_{1p} \\
1 & X_{21} & X_{22} & \dots & X_{2p} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & X_{n1} & X_{n2} & \dots & X_{np}
\end{bmatrix}
$$

donde la primera columna de unos representa el **t√©rmino intercepto** $\beta_0$.

---

## üîç Criterio de Estimaci√≥n: M√≠nimos Cuadrados Ordinarios

El objetivo del m√©todo de m√≠nimos cuadrados ordinarios es **encontrar los coeficientes $\boldsymbol{\beta}$ que minimicen la suma de los cuadrados de los errores**:

$$
S(\boldsymbol{\beta}) = \sum_{i=1}^{n} \varepsilon_i^2 = (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta})
$$

Para minimizar esta funci√≥n, derivamos con respecto a $\boldsymbol{\beta}$ y resolvemos:

$$
\frac{\partial S}{\partial \boldsymbol{\beta}} = -2\mathbf{X}^T (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta}) = 0
$$

Resolviendo para $\boldsymbol{\beta}$:

$$
\mathbf{X}^T \mathbf{X} \boldsymbol{\beta} = \mathbf{X}^T \mathbf{Y}
$$

Si la matriz $\mathbf{X}^T \mathbf{X}$ es **invertible**, podemos despejar:

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}
$$

Esta es la **soluci√≥n de m√≠nimos cuadrados ordinarios (OLS)**.

---

## üß™ Propiedades de los Estimadores OLS

Si los supuestos del modelo lineal se cumplen, los estimadores $\hat{\boldsymbol{\beta}}$ presentan las siguientes propiedades:

1. **Insesgadez**:  
   - $\mathbb{E}[\hat{\boldsymbol{\beta}}] = \boldsymbol{\beta}$  
   - En promedio, los coeficientes estimados son iguales a los valores reales.

2. **Varianza m√≠nima (Eficiencia)**:  
   - Entre todos los estimadores lineales insesgados, OLS tiene la menor varianza (propiedad de **m√≠nima varianza** dentro de los estimadores lineales insesgados, BLUE).

3. **Distribuci√≥n Asint√≥tica**:  
   - Si los errores son normales, los estimadores $\hat{\boldsymbol{\beta}}$ siguen una distribuci√≥n normal:
   $$
   \hat{\boldsymbol{\beta}} \sim \mathcal{N}(\boldsymbol{\beta}, \sigma^2 (\mathbf{X}^T \mathbf{X})^{-1})
   $$
   - Si la muestra es grande, la distribuci√≥n asint√≥tica sigue siendo normal (por el Teorema del L√≠mite Central).

---

## üî¨ Estimaci√≥n de la Varianza del Error

El **error est√°ndar de los coeficientes estimados** se obtiene a partir de la varianza de los residuos:

$$
\hat{\sigma}^2 = \frac{\sum_{i=1}^{n} \hat{\varepsilon}_i^2}{n - (p+1)}
$$

donde:

- $\hat{\varepsilon}_i = Y_i - \hat{Y}_i$ son los residuos.
- $n - (p+1)$ es el **n√∫mero de grados de libertad**, donde $p+1$ representa los coeficientes estimados (incluyendo el intercepto).

La **varianza de los coeficientes estimados** se obtiene mediante:

$$
\text{Var}(\hat{\boldsymbol{\beta}}) = \hat{\sigma}^2 (\mathbf{X}^T \mathbf{X})^{-1}
$$

---

## üìä Implementaci√≥n en Python

A continuaci√≥n, implementamos la estimaci√≥n de par√°metros usando **numpy** y **statsmodels** para validar la f√≥rmula te√≥rica.


```python
import numpy as np

# Generamos datos simulados
np.random.seed(42)
X = np.random.rand(100, 2) * 10  # Dos predictores
Y = 3 + 2.5 * X[:, 0] + 1.8 * X[:, 1] + np.random.randn(100) * 2  # Relaci√≥n lineal

# Agregamos una columna de unos para el intercepto
X_design = np.c_[np.ones(X.shape[0]), X]

# Estimaci√≥n por m√≠nimos cuadrados ordinarios (OLS)
beta_ols = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y

# Mostramos los coeficientes estimados
print(f"Coeficientes estimados (OLS): {beta_ols}")
```


### üõ†Ô∏è Validaci√≥n con `statsmodels`

Para confirmar la estimaci√≥n manual, utilizamos la biblioteca `statsmodels`.


```python

import statsmodels.api as sm

# Ajustamos un modelo con statsmodels
X_sm = sm.add_constant(X)  # Agregar intercepto
modelo = sm.OLS(Y, X_sm).fit()

# Mostramos los resultados
print(modelo.summary())
```

Para verificar la estimaci√≥n de los par√°metros, utilizamos `numpy` para la soluci√≥n manual y `statsmodels` para la validaci√≥n.


```python

import numpy as np
import pandas as pd
import statsmodels.api as sm

# Generamos datos simulados
np.random.seed(42)
X = np.random.rand(100, 2) * 10  # Dos predictores
Y = 3 + 2.5 * X[:, 0] + 1.8 * X[:, 1] + np.random.randn(100) * 2  # Relaci√≥n lineal

# Agregamos una columna de unos para el intercepto
X_design = np.c_[np.ones(X.shape[0]), X]

# Estimaci√≥n por m√≠nimos cuadrados ordinarios (OLS)
beta_ols = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y

# Mostramos los coeficientes estimados
print(f"Coeficientes estimados (OLS): {beta_ols}")
```


## üìå  Supuestos del Modelo Lineal

Para que un modelo de regresi√≥n lineal sea v√°lido, deben cumplirse los siguientes **5 supuestos fundamentales**:

1. **Linealidad**: La relaci√≥n entre las variables explicativas y la variable respuesta es lineal.
2. **Independencia**: Los errores $\varepsilon_i$ son independientes.
3. **Homoscedasticidad**: La varianza de los errores es constante en todos los valores de $X$.
4. **Normalidad**: Los errores siguen una distribuci√≥n normal.
5. **No Multicolinealidad**: No hay correlaci√≥n fuerte entre las variables explicativas.




Estos **supuestos matem√°ticos** garantizan la validez de las estimaciones y permiten realizar inferencias confiables. Estos supuestos son cr√≠ticos para que los estimadores obtenidos mediante el m√©todo de **m√≠nimos cuadrados ordinarios (OLS)** sean insesgados, eficientes y consistentes.

---

## üìå Supuesto de Linealidad

El **supuesto de linealidad** establece que la relaci√≥n entre las variables independientes $X$ y la variable dependiente $Y$ debe ser **lineal en los par√°metros**. Matem√°ticamente, el modelo de regresi√≥n lineal simple se expresa como:

$$ Y_i = \beta_0 + \beta_1 X_i + \varepsilon_i, \quad i = 1, 2, \dots, n $$

donde:

- $Y_i$ es la variable dependiente.
- $X_i$ es la variable independiente.
- $\beta_0$ y $\beta_1$ son los par√°metros del modelo.
- $\varepsilon_i$ es el t√©rmino de error, que recoge la variabilidad no explicada por $X$.

Para el caso de regresi√≥n m√∫ltiple con $p$ predictores:

$$ Y_i = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \dots + \beta_p X_{ip} + \varepsilon_i $$

El supuesto de linealidad no implica que los datos deban seguir una distribuci√≥n lineal, sino que **la relaci√≥n funcional entre $X$ y $Y$ debe ser representable mediante una combinaci√≥n lineal de los par√°metros**.

---

## üîç Evaluaci√≥n del Supuesto de Linealidad

Existen varios m√©todos para evaluar si la relaci√≥n entre las variables es lineal:

1. **An√°lisis gr√°fico**: El **gr√°fico de residuos vs. valores ajustados** es una herramienta fundamental. Si los residuos presentan **patrones sistem√°ticos** (como curvaturas), la relaci√≥n puede no ser lineal.
2. **Prueba de Ramsey RESET**: Una prueba formal que permite evaluar si el modelo omite t√©rminos no lineales significativos.



```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import reset_ramsey

sns.set(style="whitegrid")
```


### üìä An√°lisis Gr√°fico: Residuos vs. Valores Ajustados

El an√°lisis gr√°fico es una de las herramientas m√°s intuitivas para evaluar la **linealidad** del modelo. Si el modelo es correctamente especificado, los residuos deben distribuirse **aleatoriamente** alrededor de cero sin presentar patrones.

#### üö® Posibles se√±ales de no linealidad en los residuos:
- **Curvaturas**: Indican que la relaci√≥n entre $X$ y $Y$ puede ser cuadr√°tica o polin√≥mica.
- **Patrones sistem√°ticos**: Sugieren la omisi√≥n de una transformaci√≥n o variable explicativa.

A continuaci√≥n, generamos un ejemplo con una relaci√≥n cuadr√°tica para visualizar c√≥mo se comportan los residuos cuando la linealidad no se cumple.


```python

# Generaci√≥n de datos con una relaci√≥n cuadr√°tica
np.random.seed(42)
X = np.random.rand(100) * 10
Y_nolineal = 2.5 + 1.8 * X**2 + np.random.randn(100) * 5

# Ajuste de modelo lineal incorrecto
X_const = sm.add_constant(X)
modelo = sm.OLS(Y_nolineal, X_const).fit()

# Gr√°fico de residuos
plt.figure(figsize=(8, 5))
sns.scatterplot(x=modelo.fittedvalues, y=modelo.resid, color='blue', alpha=0.6)
plt.axhline(0, linestyle='dashed', color='black')
plt.xlabel("Valores Ajustados")
plt.ylabel("Residuos")
plt.title("Gr√°fico de Residuos para Evaluar Linealidad")
plt.show()
```


### üß™ Prueba de Ramsey RESET

La **prueba de Ramsey RESET** es un procedimiento estad√≠stico que eval√∫a si el modelo de regresi√≥n omite t√©rminos no lineales importantes. 

#### üìå ¬øQu√© eval√∫a esta prueba?
La prueba consiste en incluir **potencias de los valores ajustados** en el modelo y evaluar si estas contribuyen significativamente a la explicaci√≥n de la variable dependiente.

Matem√°ticamente, si nuestro modelo inicial es:

$$ Y_i = \beta_0 + \beta_1 X_i + \varepsilon_i $$

La prueba Ramsey RESET introduce t√©rminos adicionales:

$$ Y_i = \beta_0 + \beta_1 X_i + \gamma_1 \hat{Y}_i^2 + \gamma_2 \hat{Y}_i^3 + \dots + \varepsilon_i $$

donde $\hat{Y}_i$ son los valores ajustados del modelo original. La hip√≥tesis nula y alternativa son:

- $H_0$: El modelo est√° correctamente especificado (no se necesitan t√©rminos no lineales).
- $H_1$: El modelo est√° mal especificado (se requieren t√©rminos no lineales).

Si el **p-valor** es menor a **0.05**, existe evidencia suficiente para rechazar $H_0$ y concluir que el modelo no es lineal.


```python

# Prueba de Ramsey RESET para especificaci√≥n del modelo
reset_test = reset_ramsey(modelo)
print(f"Prueba de Ramsey RESET: estad√≠stico = {reset_test.statistic:.3f}, p-valor = {reset_test.pvalue:.3f}")

if reset_test.pvalue < 0.05:
    print("Se rechaza la hip√≥tesis nula. Se recomienda incluir t√©rminos no lineales.")
else:
    print("No hay evidencia suficiente para rechazar la hip√≥tesis nula. La linealidad es adecuada.")
```


### üî¨ ¬øPor qu√© funciona la prueba de Ramsey RESET?

La prueba de **Ramsey RESET (Regression Equation Specification Error Test)** se basa en la idea de que si un modelo de regresi√≥n est√° correctamente especificado, entonces los valores ajustados $\hat{Y}$ no deber√≠an contener informaci√≥n adicional que explique $Y$ m√°s all√° de lo que ya explican las variables independientes $X$. 

Matem√°ticamente, si nuestro modelo original es:

$$ Y_i = \beta_0 + \beta_1 X_i + \varepsilon_i $$

La prueba introduce t√©rminos no lineales de los valores ajustados $\hat{Y}$ en el modelo de prueba:

$$ Y_i = \beta_0 + \beta_1 X_i + \gamma_1 \hat{Y}_i^2 + \gamma_2 \hat{Y}_i^3 + \dots + \varepsilon_i $$

donde $\hat{Y}_i$ son los valores ajustados del modelo original. 

Si el modelo original es correcto, los t√©rminos adicionales $\gamma_1$ y $\gamma_2$ deber√≠an ser **estad√≠sticamente no significativos**, ya que no deber√≠an explicar variabilidad adicional en $Y$.

#### üìå ¬øC√≥mo se eval√∫a?
- Se realiza una **prueba F** para comparar el modelo original con el modelo extendido.
- Se eval√∫a si agregar t√©rminos polin√≥micos mejora significativamente el ajuste del modelo.

#### üìå Interpretaci√≥n:
- **Si el p-valor es mayor a 0.05**: No hay evidencia suficiente para rechazar la hip√≥tesis nula, lo que sugiere que el modelo es adecuado.
- **Si el p-valor es menor a 0.05**: Se rechaza la hip√≥tesis nula, lo que indica que el modelo podr√≠a estar mal especificado y requerir t√©rminos no lineales.

---

A continuaci√≥n, aplicamos la prueba de Ramsey RESET a nuestro modelo.


```python

# Prueba de Ramsey RESET para especificaci√≥n del modelo
reset_test = reset_ramsey(modelo)
print(f"Prueba de Ramsey RESET: estad√≠stico = {reset_test.statistic:.3f}, p-valor = {reset_test.pvalue:.3f}")

if reset_test.pvalue < 0.05:
    print("Se rechaza la hip√≥tesis nula. Se recomienda incluir t√©rminos no lineales.")
else:
    print("No hay evidencia suficiente para rechazar la hip√≥tesis nula. La linealidad es adecuada.")
```

# üìå Supuesto de Independencia de los Errores

El **supuesto de independencia** establece que los errores $\varepsilon_i$ de un modelo de regresi√≥n deben ser **mutuamente independientes**. En otras palabras, el valor del error en una observaci√≥n no debe estar correlacionado con el valor del error en otra observaci√≥n.

Matem√°ticamente, esto se expresa como:

$$ \text{Cov}(\varepsilon_i, \varepsilon_j) = 0, \quad \forall i \neq j $$

donde $\text{Cov}$ representa la covarianza entre los errores. Si esta covarianza es distinta de cero, hay **dependencia en los errores**, lo que puede indicar la presencia de **autocorrelaci√≥n**.

---

## ‚úÖ Importancia del Supuesto

El incumplimiento del supuesto de independencia puede llevar a:

- **Inferencias err√≥neas**: Si los errores est√°n correlacionados, las pruebas de hip√≥tesis y los intervalos de confianza pueden ser incorrectos.
- **Coeficientes ineficientes**: Aunque los estimadores de M√≠nimos Cuadrados Ordinarios (OLS) siguen siendo insesgados, dejan de ser eficientes y pueden tener varianzas m√°s altas de lo esperado.
- **Predicciones sesgadas**: Un modelo con errores correlacionados puede generar predicciones incorrectas, especialmente en datos secuenciales o temporales.

---

## üîç M√©todos de Detecci√≥n

Para evaluar si los errores son independientes, utilizamos las siguientes herramientas:

1. **An√°lisis gr√°fico**: El **gr√°fico de residuos en funci√≥n del tiempo o del √≠ndice** permite identificar patrones en los residuos.
2. **Prueba de Durbin-Watson**: Es una prueba estad√≠stica dise√±ada para detectar **autocorrelaci√≥n de primer orden** en los errores.



```python

# Generaci√≥n de datos con autocorrelaci√≥n en los errores
np.random.seed(42)
X_dep = np.random.rand(100) * 10
Y_dependencia = 2.5 + 1.8 * X_dep + np.cumsum(np.random.randn(100))  # Ruido acumulado genera dependencia

# Ajuste del modelo
X_dep_const = sm.add_constant(X_dep)
modelo_dep = sm.OLS(Y_dependencia, X_dep_const).fit()

# Gr√°fico de residuos en funci√≥n del √≠ndice
plt.figure(figsize=(8, 5))
plt.plot(modelo_dep.resid, marker='o', linestyle='-', color='red', alpha=0.7)
plt.axhline(0, linestyle='dashed', color='black')
plt.xlabel("√çndice de la observaci√≥n")
plt.ylabel("Residuos")
plt.title("Gr√°fico de Residuos para Evaluar Independencia")
plt.show()
```


### üß™ Prueba de Durbin-Watson

La **prueba de Durbin-Watson** eval√∫a la presencia de **autocorrelaci√≥n de primer orden** en los errores del modelo. 

#### üìå ¬øQu√© eval√∫a esta prueba?

La autocorrelaci√≥n de primer orden se define como:

$$ \rho = \frac{\sum_{i=2}^{n} (\varepsilon_i - \varepsilon_{i-1})^2}{\sum_{i=1}^{n} \varepsilon_i^2} $$

Donde $\rho$ representa el grado de correlaci√≥n entre los errores consecutivos.

La estad√≠stica de Durbin-Watson ($DW$) se calcula como:

$$ DW = 2(1 - \rho) $$

#### üìå Interpretaci√≥n:

- **$DW \approx 2$**: No hay autocorrelaci√≥n en los errores (modelo adecuado).
- **$DW < 1.5$**: Hay autocorrelaci√≥n positiva (los errores consecutivos tienden a ser similares).
- **$DW > 2.5$**: Hay autocorrelaci√≥n negativa (los errores consecutivos tienden a alternar signos).

---

A continuaci√≥n, aplicamos la prueba de Durbin-Watson a nuestro modelo.


```python
from statsmodels.stats.stattools import durbin_watson

# Prueba de Durbin-Watson para independencia de los errores
dw_stat_dep = durbin_watson(modelo_dep.resid)
print(f"Estad√≠stico Durbin-Watson: {dw_stat_dep:.3f} (Cerca de 2 indica independencia)")

if dw_stat_dep < 1.5:
    print("Posible autocorrelaci√≥n positiva en los errores.")
elif dw_stat_dep > 2.5:
    print("Posible autocorrelaci√≥n negativa en los errores.")
else:
    print("No hay evidencia de autocorrelaci√≥n en los errores.")
```


### üîß M√©todos para Corregir la Autocorrelaci√≥n

Si se detecta autocorrelaci√≥n en los errores, existen diversas estrategias para corregir el problema:

1. **Modelos de series temporales**:
   - Si los datos presentan una estructura temporal, utilizar modelos como **ARIMA** o **regresi√≥n con t√©rminos autorregresivos**.

2. **Incluir variables omitidas**:
   - A veces, la autocorrelaci√≥n se debe a la omisi√≥n de una variable explicativa clave que no ha sido incluida en el modelo.

3. **Regresi√≥n con Errores Estructurados**:
   - Aplicar modelos con **errores heteroced√°sticos y autorregresivos (HAC)** mediante t√©cnicas como la **correcci√≥n de Newey-West**.

# üìå Supuesto de Homocedasticidad

El **supuesto de homocedasticidad** establece que la varianza de los errores $\varepsilon_i$ es **constante** en todos los niveles de las variables explicativas. Matem√°ticamente, se expresa como:

$$ \text{Var}(\varepsilon_i) = \sigma^2, \quad \forall i $$

Si este supuesto no se cumple y la varianza de los errores var√≠a con los valores de $X$, se dice que el modelo presenta **heterocedasticidad**.

---

## ‚úÖ Importancia del Supuesto

La homocedasticidad es fundamental para que los **intervalos de confianza y pruebas de hip√≥tesis sean correctos**. Si hay heterocedasticidad:

- Los estimadores de m√≠nimos cuadrados ordinarios (OLS) siguen siendo insesgados, pero **pierden eficiencia**, lo que afecta la precisi√≥n de las predicciones.
- Las pruebas estad√≠sticas pueden arrojar resultados incorrectos porque los errores est√°ndar de los coeficientes pueden estar mal estimados.
- El modelo puede dar m√°s peso a ciertas observaciones, afectando la interpretaci√≥n de los coeficientes.

---

## üîç M√©todos de Detecci√≥n

Para evaluar si los errores son homoced√°sticos, utilizamos las siguientes herramientas:

1. **Gr√°fico de residuos vs. valores ajustados**: Si la dispersi√≥n de los residuos cambia a medida que aumentan los valores ajustados, hay indicios de heterocedasticidad.
2. **Prueba de Breusch-Pagan**: Eval√∫a si la varianza de los errores depende de los valores de las variables explicativas.



```python

# Generaci√≥n de datos con heterocedasticidad
np.random.seed(42)
X_hetero = np.random.rand(100) * 10
Y_hetero = 2.5 + 1.8 * X_hetero + (X_hetero * np.random.randn(100))  # Errores aumentan con X

# Ajuste del modelo
X_hetero_const = sm.add_constant(X_hetero)
modelo_hetero = sm.OLS(Y_hetero, X_hetero_const).fit()

# Gr√°fico de residuos
plt.figure(figsize=(8, 5))
sns.scatterplot(x=modelo_hetero.fittedvalues, y=modelo_hetero.resid, color='green', alpha=0.6)
plt.axhline(0, linestyle='dashed', color='black')
plt.xlabel("Valores Ajustados")
plt.ylabel("Residuos")
plt.title("Gr√°fico de Residuos para Evaluar Homocedasticidad")
plt.show()
```


### üß™ Prueba de Breusch-Pagan

La **prueba de Breusch-Pagan** eval√∫a si la varianza de los errores est√° relacionada con los valores de las variables explicativas.

#### üìå ¬øQu√© eval√∫a esta prueba?

Si la varianza de los errores depende de $X$, entonces existe heterocedasticidad. La prueba se basa en la siguiente regresi√≥n auxiliar:

$$ \hat{\varepsilon}^2 = \alpha_0 + \alpha_1 X_1 + \alpha_2 X_2 + \dots + \alpha_p X_p + \nu $$

Se realiza una **prueba de chi-cuadrado** para determinar si los coeficientes $\alpha_i$ son significativamente diferentes de cero.

#### üìå Hip√≥tesis de la prueba:

- **$H_0$**: La varianza de los errores es constante (homocedasticidad).
- **$H_1$**: La varianza de los errores depende de $X$ (heterocedasticidad).

Si el **p-valor** es menor a **0.05**, se rechaza la hip√≥tesis nula y se concluye que el modelo presenta heterocedasticidad.


```python
from statsmodels.stats.diagnostic import het_breuschpagan

# Prueba de Breusch-Pagan para homocedasticidad
_, pval_bp, _, _ = het_breuschpagan(modelo_hetero.resid, X_hetero_const)
print(f"Prueba de Breusch-Pagan: p-valor = {pval_bp:.3f} (p > 0.05 indica varianza constante)")

if pval_bp < 0.05:
    print("Se detecta heterocedasticidad en el modelo.")
else:
    print("No hay evidencia suficiente de heterocedasticidad.")
```


### üîß M√©todos para Corregir la Heterocedasticidad

Si la heterocedasticidad es detectada en el modelo, se pueden emplear diversas estrategias para corregirla:

1. **Transformaciones de la variable dependiente**:
   - Aplicar logaritmos o ra√≠ces cuadradas a la variable $Y$ puede estabilizar la varianza.

2. **Regresi√≥n ponderada**:
   - Dar m√°s peso a las observaciones con menor varianza y menos peso a las observaciones con mayor varianza.

3. **Modelos de estimaci√≥n robusta**:
   - Usar errores est√°ndar robustos de **White** para ajustar las pruebas de hip√≥tesis y obtener estimaciones correctas.

4. **Modelos de regresi√≥n generalizada**:
   - En modelos m√°s avanzados, se pueden utilizar modelos como **GLS (Generalized Least Squares)** para corregir heterocedasticidad expl√≠citamente.

# üìå Supuesto de Normalidad de los Errores

El **supuesto de normalidad de los errores** establece que los residuos del modelo de regresi√≥n deben seguir una distribuci√≥n normal con media cero y varianza constante. Matem√°ticamente:

$$ \varepsilon_i \sim \mathcal{N}(0, \sigma^2) $$

donde $\mathcal{N}(0, \sigma^2)$ denota una distribuci√≥n normal con media $0$ y varianza $\sigma^2$.

Este supuesto es fundamental para realizar **pruebas de hip√≥tesis v√°lidas** sobre los coeficientes del modelo.

---

## ‚úÖ Importancia del Supuesto

- **Inferencias precisas**: La normalidad de los errores garantiza que las pruebas estad√≠sticas sobre los coeficientes sean v√°lidas.
- **Correcta construcci√≥n de intervalos de confianza**: Si los errores no son normales, los intervalos de confianza y los valores p pueden estar mal estimados.
- **Modelos con supuestos fuertes**: Algunos m√©todos avanzados, como la regresi√≥n Bayesiana, requieren supuestos m√°s estrictos sobre la distribuci√≥n de los errores.

Si el tama√±o muestral es grande, el **Teorema del L√≠mite Central** permite que las estimaciones sean v√°lidas a√∫n si los errores no son perfectamente normales.

---

## üîç M√©todos de Detecci√≥n

Para evaluar si los errores siguen una distribuci√≥n normal, utilizamos las siguientes herramientas:

1. **Histograma y Gr√°fico de Densidad**: Nos permite inspeccionar visualmente si los residuos tienen forma de campana.
2. **Gr√°fico Q-Q (Quantile-Quantile)**: Compara la distribuci√≥n emp√≠rica de los errores con una distribuci√≥n normal te√≥rica.
3. **Prueba de Shapiro-Wilk**: Una prueba estad√≠stica para verificar si los errores siguen una distribuci√≥n normal.



```python

# Generaci√≥n de datos con errores no normales
np.random.seed(42)
X_nonnormal = np.random.rand(100) * 10
Y_nonnormal = 2.5 + 1.8 * X_nonnormal + np.random.laplace(0, 3, 100)  # Errores con distribuci√≥n Laplace

# Ajuste del modelo
X_nonnormal_const = sm.add_constant(X_nonnormal)
modelo_nonnormal = sm.OLS(Y_nonnormal, X_nonnormal_const).fit()

# Histograma de los residuos
plt.figure(figsize=(8, 5))
sns.histplot(modelo_nonnormal.resid, kde=True, color='purple', bins=15)
plt.xlabel("Residuos")
plt.title("Distribuci√≥n de los Residuos (Evaluaci√≥n de Normalidad)")
plt.show()
```


### üìä Gr√°fico Q-Q (Quantile-Quantile)

El **gr√°fico Q-Q** compara los cuantiles de los residuos con los cuantiles de una distribuci√≥n normal te√≥rica. Si los residuos siguen una distribuci√≥n normal, los puntos deben estar alineados sobre la diagonal.



```python

# Gr√°fico Q-Q para evaluar normalidad
import scipy.stats as stats

plt.figure(figsize=(6, 6))
stats.probplot(modelo_nonnormal.resid, dist="norm", plot=plt)
plt.title("Gr√°fico Q-Q de los Residuos")
plt.show()
```


### üß™ Prueba de Shapiro-Wilk

La **prueba de Shapiro-Wilk** eval√∫a la hip√≥tesis de que los residuos siguen una distribuci√≥n normal.

#### üìå ¬øQu√© eval√∫a esta prueba?

Se calcula el coeficiente $W$ que mide la correlaci√≥n entre los valores observados y los valores esperados de una distribuci√≥n normal.

#### üìå Hip√≥tesis de la prueba:

- **$H_0$**: Los errores siguen una distribuci√≥n normal.
- **$H_1$**: Los errores no siguen una distribuci√≥n normal.

Si el **p-valor** es menor a **0.05**, se rechaza la hip√≥tesis nula y se concluye que los errores no siguen una distribuci√≥n normal.


```python
from scipy.stats import shapiro

# Prueba de Shapiro-Wilk para normalidad
shapiro_stat, shapiro_pval = shapiro(modelo_nonnormal.resid)
print(f"Prueba de Shapiro-Wilk: p-valor = {shapiro_pval:.3f} (p > 0.05 indica normalidad)")

if shapiro_pval < 0.05:
    print("Se detecta no normalidad en los errores.")
else:
    print("No hay evidencia suficiente para rechazar la normalidad.")
```


### üîß M√©todos para Corregir la No Normalidad

Si se detecta que los errores no son normales, existen diversas estrategias para mitigar el problema:

1. **Transformaciones de la variable dependiente**:
   - Aplicar **logaritmos**, ra√≠z cuadrada o transformaciones de Box-Cox para modificar la distribuci√≥n de los residuos.

2. **Uso de estimadores robustos**:
   - Aplicar m√©todos como **regresi√≥n robusta (Huber, RANSAC)** que no dependen fuertemente de la normalidad de los errores.

3. **Aumento del tama√±o de la muestra**:
   - Si el tama√±o muestral es peque√±o, la normalidad de los errores puede no observarse. Al aumentar la muestra, el **Teorema del L√≠mite Central** asegura que las inferencias sigan siendo v√°lidas.

4. **Usar modelos alternativos**:
   - Si los residuos siguen una distribuci√≥n espec√≠fica distinta de la normal, considerar modelos como **regresi√≥n cuant√≠lica** o **modelos de distribuci√≥n flexible**.

# üìå Supuesto de No Multicolinealidad

El **supuesto de no multicolinealidad** establece que las variables explicativas en un modelo de regresi√≥n no deben estar altamente correlacionadas entre s√≠. Si existe una fuerte relaci√≥n lineal entre dos o m√°s predictores, se produce **multicolinealidad**, lo que dificulta la estimaci√≥n precisa de los coeficientes.

Matem√°ticamente, si una variable explicativa $X_j$ puede expresarse aproximadamente como una combinaci√≥n lineal de otras variables:

$$ X_j = \alpha_1 X_1 + \alpha_2 X_2 + \dots + \alpha_p X_p + \varepsilon $$

entonces existe **colinealidad**.

---

## ‚úÖ Importancia del Supuesto

Si las variables predictoras est√°n altamente correlacionadas:

- **Los coeficientes de regresi√≥n se vuelven inestables**: Peque√±as variaciones en los datos pueden causar grandes cambios en las estimaciones de los coeficientes.
- **Dificultad en la interpretaci√≥n**: Es dif√≠cil determinar el impacto individual de cada variable en la variable dependiente.
- **Aumento de la varianza de los coeficientes estimados**: Lo que reduce la precisi√≥n del modelo.

---

## üîç M√©todos de Detecci√≥n

Para evaluar si el modelo presenta multicolinealidad, utilizamos:

1. **Matriz de correlaci√≥n**: Si dos variables tienen una correlaci√≥n mayor a **0.8 o 0.9**, hay riesgo de multicolinealidad.
2. **Factor de Inflaci√≥n de la Varianza (VIF)**: Un **VIF mayor a 10** indica una correlaci√≥n muy alta entre variables predictoras.



```python

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Generar datos con alta correlaci√≥n
np.random.seed(42)
X1 = np.random.rand(100) * 10
X2 = X1 + np.random.randn(100) * 0.1  # Segunda variable altamente correlacionada con X1
X3 = np.random.rand(100) * 10  # Variable independiente

# Crear DataFrame
df_multicolinealidad = pd.DataFrame({"X1": X1, "X2": X2, "X3": X3})

# Calcular la matriz de correlaci√≥n
correlation_matrix = df_multicolinealidad.corr()
print("Matriz de correlaci√≥n:")
print(correlation_matrix)

# Calcular el VIF para cada variable
vif_data = pd.DataFrame()
vif_data["Variable"] = df_multicolinealidad.columns
vif_data["VIF"] = [variance_inflation_factor(df_multicolinealidad.values, i) for i in range(df_multicolinealidad.shape[1])]

print("Factor de Inflaci√≥n de la Varianza (VIF):")
print(vif_data)
```


### üß™ Interpretaci√≥n del Factor de Inflaci√≥n de la Varianza (VIF)

El **VIF** mide cu√°nto aumenta la varianza de un coeficiente debido a la correlaci√≥n con otras variables.

#### üìå Interpretaci√≥n:
- **VIF $\approx 1$**: No hay multicolinealidad.
- **VIF entre 5 y 10**: Multicolinealidad moderada (posible problema).
- **VIF > 10**: Multicolinealidad severa (problema grave).

Si el VIF es alto, significa que la variable est√° fuertemente correlacionada con otras variables y puede afectar la estabilidad del modelo.




### üîß M√©todos para Corregir la Multicolinealidad

Si se detecta multicolinealidad en el modelo, existen varias estrategias para corregir el problema:

1. **Eliminar una de las variables correlacionadas**:
   - Si dos variables tienen una correlaci√≥n superior a **0.9**, considerar eliminar una de ellas.

2. **Usar t√©cnicas de reducci√≥n de dimensionalidad**:
   - Aplicar **An√°lisis de Componentes Principales (PCA)** para reducir la redundancia en los datos.

3. **Transformaciones matem√°ticas**:
   - Si las variables representan informaci√≥n similar, considerar su combinaci√≥n en una √∫nica variable sint√©tica.

4. **Usar regularizaci√≥n (Ridge Regression o Lasso)**:
   - Estos m√©todos penalizan coeficientes grandes y reducen el impacto de la multicolinealidad.



# üìå Diagn√≥stico Autom√°tico de Supuestos en Modelos Lineales

Este cuaderno proporciona una funci√≥n en Python que eval√∫a autom√°ticamente los **5 supuestos fundamentales del modelo de regresi√≥n lineal**:

‚úÖ **Supuesto de Linealidad**: Gr√°fico de residuos vs. valores ajustados y prueba de Ramsey RESET.  
‚úÖ **Supuesto de Independencia**: Prueba de **Durbin-Watson** para detectar autocorrelaci√≥n en los errores.  
‚úÖ **Supuesto de Homocedasticidad**: Prueba de **Breusch-Pagan** para detectar cambios en la varianza de los errores.  
‚úÖ **Supuesto de Normalidad**: Histograma de residuos, gr√°fico Q-Q y prueba de **Shapiro-Wilk**.  
‚úÖ **Supuesto de No Multicolinealidad**: Matriz de correlaci√≥n y **Factor de Inflaci√≥n de la Varianza (VIF)**.


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.outliers_influence import reset_ramsey, variance_inflation_factor
from scipy.stats import shapiro
from scipy import stats

sns.set(style="whitegrid")
```

```python

def diagnostico_modelo(modelo, X, Y):
    """
    Eval√∫a los cinco supuestos del modelo de regresi√≥n lineal.
    
    Par√°metros:
    modelo : statsmodels.regression.linear_model.RegressionResultsWrapper
        Modelo ajustado con statsmodels OLS.
    X : ndarray o DataFrame
        Matriz de predictores (sin la columna de unos para el intercepto).
    Y : ndarray o Series
        Vector de valores observados de la variable dependiente.
    """
    
    residuos = modelo.resid
    valores_ajustados = modelo.fittedvalues

    print("üîç Evaluaci√≥n de Supuestos del Modelo Lineal")

    # 1. Supuesto de Linealidad
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=valores_ajustados, y=residuos, color='blue', alpha=0.6)
    plt.axhline(0, linestyle='dashed', color='black')
    plt.xlabel("Valores Ajustados")
    plt.ylabel("Residuos")
    plt.title("Gr√°fico de Residuos vs. Valores Ajustados (Evaluaci√≥n de Linealidad)")
    plt.show()

    # Prueba de Ramsey RESET
    reset_test = reset_ramsey(modelo)
    print(f"üìå Prueba de Ramsey RESET: estad√≠stico = {reset_test.statistic:.3f}, p-valor = {reset_test.pvalue:.3f}")
    if reset_test.pvalue < 0.05:
        print("‚ùå Se recomienda incluir t√©rminos no lineales en el modelo.")
    else:
        print("‚úÖ No hay evidencia de no linealidad en el modelo.")

    # 2. Supuesto de Independencia
    dw_stat = durbin_watson(residuos)
    print(f"üìå Prueba de Durbin-Watson: {dw_stat:.3f} (Cerca de 2 indica independencia)")
    if dw_stat < 1.5:
        print("‚ùå Posible autocorrelaci√≥n positiva en los errores.")
    elif dw_stat > 2.5:
        print("‚ùå Posible autocorrelaci√≥n negativa en los errores.")
    else:
        print("‚úÖ No hay evidencia de autocorrelaci√≥n en los errores.")

    # 3. Supuesto de Homocedasticidad
    _, pval_bp, _, _ = het_breuschpagan(residuos, modelo.model.exog)
    print(f"üìå Prueba de Breusch-Pagan: p-valor = {pval_bp:.3f} (p > 0.05 indica varianza constante)")
    if pval_bp < 0.05:
        print("‚ùå Se detecta heterocedasticidad en el modelo.")
    else:
        print("‚úÖ No hay evidencia suficiente de heterocedasticidad.")

    # 4. Supuesto de Normalidad
    plt.figure(figsize=(8, 5))
    sns.histplot(residuos, kde=True, color='purple', bins=15)
    plt.xlabel("Residuos")
    plt.title("Distribuci√≥n de los Residuos (Evaluaci√≥n de Normalidad)")
    plt.show()

    plt.figure(figsize=(6, 6))
    stats.probplot(residuos, dist="norm", plot=plt)
    plt.title("Gr√°fico Q-Q de los Residuos")
    plt.show()

    shapiro_stat, shapiro_pval = shapiro(residuos)
    print(f"üìå Prueba de Shapiro-Wilk: p-valor = {shapiro_pval:.3f} (p > 0.05 indica normalidad)")
    if shapiro_pval < 0.05:
        print("‚ùå Se detecta no normalidad en los errores.")
    else:
        print("‚úÖ No hay evidencia suficiente para rechazar la normalidad.")

    # 5. Supuesto de No Multicolinealidad
    vif_data = pd.DataFrame()
    vif_data["Variable"] = [f"X{i}" for i in range(1, X.shape[1] + 1)]
    vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

    print("üìå Factor de Inflaci√≥n de la Varianza (VIF):")
    print(vif_data)
    if vif_data["VIF"].max() > 10:
        print("‚ùå Se detecta un problema de multicolinealidad en el modelo.")

    print("‚úÖ Diagn√≥stico Completo")
```


## üìä Aplicaci√≥n de la Funci√≥n

Probemos la funci√≥n de diagn√≥stico en un modelo de regresi√≥n ajustado con `statsmodels`.


```python

# Generaci√≥n de datos simulados
np.random.seed(42)
X_test = np.random.rand(100, 2) * 10  # Dos predictores
Y_test = 3 + 2.5 * X_test[:, 0] + 1.8 * X_test[:, 1] + np.random.randn(100) * 2  # Relaci√≥n lineal

# Ajustamos un modelo con statsmodels
X_sm_test = sm.add_constant(X_test)  # Agregar intercepto
modelo_test = sm.OLS(Y_test, X_sm_test).fit()

# Ejecutamos el diagn√≥stico autom√°tico
diagnostico_modelo(modelo_test, X_test, Y_test)
```
